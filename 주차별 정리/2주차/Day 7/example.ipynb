# 막대 그래프 (Bar Chart)

- '범주형 데이터의 빈도나 값'을 나타내기 위해 사용되는 기본적인 시각화 도구
- 데이터 값은 수평 또는 수직 막대로 표현되며, 각 막대의 길이는 해당 범주의 크기나 빈도를 나타냄
- 카테고리별 지표값을 절대적으로도 상대적으로도 정확하게 표현할 수 있다(축은 0부터 시작되는 것이 중요함).
- 응용 버전으로는 막대 그래프 (그라데이션), 적층 막대 그래프, 100% 적층 막대 그래프, 블래드 차트 등이 있음
- 예시)
  - 비지니스: 월별 매출 비교를 통한 성과 분석
  - 교육: 학년별 성적 비교
  - 의료: 병원별 환자 수 비교

```python
# 데이터 정의
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [150, 200, 180, 220, 250, 230]
```

```python
# 막대그래프 생성
plt.figure(figsize=(10, 6)) # 그래프 크기 설정
plt.bar(months, sales, color='skyblue', label='Sales') # 막대그래프 생성
plt.title('Monthly Sales') # 제목 추가
plt.xlabel('Month') # x축 레이블 추가
plt.ylabel('Sales') # y축 레이블 추가
plt.legend() # 범례 추가
plt.grid(True) # 그리드 추가
plt.show() # 그래프 출력
```

# 파이 차트 (Pie Chart)

- 전체에서 각 부분의 비율을 시각화하는 데 사용
- 원형 차트로, 각 조각의 크기는 저네에서 해당 항목의 비율을 나타냄
- 다수의 카테고리로 구성하는 것은 효과적이지 않고 카테고리가 4개 이상이면 막대 그래프로 대체하는 것이 바람직함
- 파생 유형으로는 도넛 차트가 있음(원 그래프의 중심을 도려내고 레이블 표시장소로 활용해 공간을 유효하게 활용 가능)
- 예시)
  - 비지니스: 예산 분배, 시장 점유율 비교
  - 교육: 학과별 학생 비율
  - 의료: 병원 내 각 부서별 환자 비율

```python
# 데이터 정의
labels = ['Product A', 'Product B', 'Product C', 'Product D']
sizes = [400, 300, 500, 100]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0) # 첫 번째 조각을 약간 떨어뜨리기
```

```python
# 간결한 디자인의 파이차트
plt.figure(figsize=(8, 8)) # 그래프 크기 설정
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution - Simple Design')
plt.axis('equal') # 파이차트를 원형으로 유지
plt.show()
```

```python
# 복잡한 디자인의 파이차트
plt.figure(figsize=(8, 8)) # 그래프 크기 설정
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Sales Distribution - Complex Design')
plt.axis('equal')
plt.show()
```

# 선그래프 (Line Chart)

- 데이터 포인트가 선으로 연결되어 시간 경과에 따른 추세를 나타냄
- 가로축의 연속적인 값에서 세로축의 수량을 비교하기 위해 사용
- 막대 그래프와는 달리 상대적인 경향만 표현할 때는 세로축이 0으로 시작하지 않아도 됨
- 응용 버전으로는 면 그래프, 적립 면 그래프, 100% 적립 면 그래프, 슬로프 차트, 반프 차트, 스파크 차트 등이 있음
- 예시)
  - 비지니스: 주간 매출 트렌드 분석
  - 교육: 학기별 학생 출석률 변화
  - 건강: 일별 혈압 수치 변화

```python
# 데이터 정의
data = {'Year': [2018, 2019, 2020, 2021], 'Visitors': [12000, 18000, 23000, 30000]}
df = pd.DataFrame(data)
```

```python
# 선그래프 생성
plt.figure(figsize=(10, 6)) # 그래프 크기 설정
sns.lineplot(data=df, x='Year', y='Visitors', marker='o', label='Visitors')
plt.title('Annual Website Visitors') # 제목 추가
plt.xlabel('Year') # x축 레이블 추가
plt.ylabel('Number of Visitors') # y축 레이블 추가
plt.legend() # 범례 추가
plt.grid(True) # 그리드 추가
plt.show()
```

```python
# 데이터 정의
data = {'Year': [2018, 2019, 2020, 2021],
        'Product A': [1500, 2000, 2400, 2800],
        'Product B': [2300, 2700, 3000, 3500],
        'Product C': [1800, 2200, 2600, 3000]}

df = pd.DataFrame(data)
```

```python
# 복합 차트 생성
plt.figure(figsize=(10, 6)) # 그래프 크기 설정
sns.lineplot(x='Year', y='Product A', data=df, marker='o', label='Product A')
sns.lineplot(x='Year', y='Product B', data=df, marker='o', label='Product B')
sns.lineplot(x='Year', y='Product C', data=df, marker='o', label='Product C')
plt.title('Annual Product Sales') # 제목 추가
plt.xlabel('Year') # x축 레이블 추가
plt.ylabel('Sales') # y축 레이블 추ㅏ
plt.legend() # 범례 추가
plt.grid(True) # 그리드 추가
plt.show()
```

# 히스토그램 (Histogram)

- '연속형 데이터의 분포'를 나타내는 데 사용
- 데이터 범위를 여러 구간(bin)으로 나누고 각 구간의 데이터 빈도를 막대로 표시
- 예시)
  - 비지니스: 제품 가격대별 판매량 분포
  - 교육: 시험 점수 분포
  - 건강: 나이대별 환자 수 분포

```python
# 데이터 생성
np.random.seed(42)
data = np.random.randn(1000)
```

```python
plt.figure(figsize=(10, 6)) # 그래프 크기 설정
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Random Data') # 제목 추가
plt.xlabel('Value') # x축 레이블 추가
plt.ylabel('Frequency') # y축 레이블 추가
plt.grid(True) # 그리드 추가
plt.show()
```

# 산점도 (Scatter Plot)

- 두 변수 간의 관계를 시각화하는 데 사용
- 각 데이터 포인트는 두 변수 값의 교차점에 위치
- X축과 Y축울 플롯, 패턴이나 상관, 클러스터, 아웃라이어를 나타냄
- 응용 버전으로는 버블 차트, 산포도 매트릭스 등이 있음
- 예시)
  - 비지니스: 광고비와 매출 간의 상관관계 분석
  - 교육: 공부 시간과 성적 간의 관계
  - 건강: 체중과 혈압 간의 관계

```python
# 데이터 생성
np.random.seed(42)
data = {
    'x': np.random.randn(100),
    'y': np.random.randn(100)
}
df = pd.DataFrame(data)
```

```python
# 산점도 생성
plt.figure(figsize=(10, 6)) # 그래프 크기 설정
sns.scatterplot(data=df, x='x', y='y')
plt.title('Scatter Plot of Random Data') # 제목 추가
plt.xlabel('X') # x축 레이블 추가
plt.ylabel('Y') # y축 레이블 추가
plt.grid(True) # 그리드 추가
plt.show()
```

# 박스플롯 (Box Plot)

- 데이터의 분포와 이상치를 시각화하는 데 사용
- 중앙값, 사분위수, 최소값, 최대값 및 이상치를 표시
- 편차 분석에 용이
- 예시)
  - 비지니스: 직원 연봉 분포 분석
  - 교육: 시험 점수 분포와 이상치 분석
  - 건강: 체질량지수(BMI) 분포

```python
# 데이터 생성
np.random.seed(42)
data = {
    'Category': np.random.choice(['A', 'B', 'C'], 100),
    'Value': np.random.randn(100)
}
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 6)) # 그래프 크기 설정
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Box Plot of Random Data by Category') # 제목 추가
plt.xlabel('Category') # x축 레이블 추가
plt.ylabel('Value') # y축 레이블 추가
plt.grid(True) # 그리드 추가
plt.show()
```
