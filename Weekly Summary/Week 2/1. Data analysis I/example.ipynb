```python
def print_separator(length: int=40) -> None:
    print('-' * length)
```

```python
import pandas as pd

# Load Dataset
data = pd.read_csv('./HeartDiseaseTrain-Test.csv')
 
print_separator()
print("데이터의 처음 몇 줄과 각 칼럼에 대한 정보 확인")

# Displaying the first few rows
display(data.head())

# Displaying the data info
data_info = data.info()

# Using display to show a summary
display(data.describe(include='all'))
```

```python
print_separator()
print("Data Types:\n", data.dtypes)

categorical_cols = data.select_dtypes(include=['object', 'category']).columns
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns

print_separator()
print("Categorical Columns:\n", categorical_cols)

print_separator()
print("Numerical Columns:\n", numerical_cols)
```

```python
print_separator()
print("Missing Values:\n", data.isnull().sum())

print_separator()
for col in categorical_cols:
    print(f"Unique values in {col}:\n{data[col].value_counts()}\n")
    
print_separator()
print("Descriptive Statistics for Numerical Data:\n", data[numerical_cols].describe())
```

왜도(Skewness)
- 0에 가까울수록 정규분포에 근사
- 양의 값은 오른쪽 꼬리가 긴 분포(왼쪽으로 치우친), 음의 값은 왼쪽 꼬리가 긴 분포(오른쪽으로 치우친)

첨도(Kurtosis)
- 0에 가까울수록 정규분포에 근사
- 값이 크면 뾰족하고, 작으면 평평

```python
print_separator()
print("Skewness of the data:\n", data[numerical_cols].skew())

print_separator()
print("\nKurtosis of the data:\n", data[numerical_cols].kurt())
```

```python
data['resting_blood_pressure'].plot(kind='kde')
```

상관계수 값이 1에 가까울수록 완벽한 양의 상관관계, -1에 가까울수록 완벽한 음의 상관관계를 나타냄

```python
print_separator()
print("Pearson Correlation:\n", data[numerical_cols].corr(method='pearson'))

print_separator()
print("Spearman Correlation:\n", data[numerical_cols].corr(method='spearman'))
```

T-통계량의 절대값이 클 수록 두 그룹 간의 차이가 크다고 할 수 있음

일반적으로 P-value가 0.05보다 작으면 귀무 가설을 기각하고, 통계적으로 유의미한 차이가 있음을 인정함

```python
from scipy import stats

# 남성과 여성 그룹 데이터 분할
male_max_hr = data[data['sex'] == 'Male']['Max_heart_rate']
female_max_hr = data[data['sex'] == 'Female']['Max_heart_rate']

# 두 그룹 간의 평균 최대 심박수 차이 검정
t_stat, p_val = stats.ttest_ind(male_max_hr, female_max_hr)
print_separator()
print(f"T-statistics: {t_stat}, P-value: {p_val}")
```
